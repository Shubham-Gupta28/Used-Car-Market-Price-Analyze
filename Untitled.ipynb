```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
import numpy as np
import time
import re
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_squared_error, r2_score
import xgboost as xgb
import random
```

```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.chrome.service import Service
from webdriver_manager.chrome import ChromeDriverManager
from selenium.webdriver.chrome.options import Options
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from selenium.common.exceptions import NoSuchElementException, TimeoutException, StaleElementReferenceException
```

```python
options = Options()
options.add_argument('--headless')
options.add_argument('--no-sandbox')
options.add_argument('--disable-dev-shm-usage')
options.add_argument('--disable-gpu')
options.add_argument('--window-size=1920,1080')
options.add_argument('user-agent=Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/120.0.0.0 Safari/537.36')

driver = webdriver.Chrome(service=Service(ChromeDriverManager().install()), options=options)


# Car brands list
CAR_BRANDS = [
    'Maruti Suzuki', 'Maruti', 'Suzuki', 'Hyundai', 'Honda', 'Toyota', 
    'Mahindra', 'Tata', 'Kia', 'MG', 'Renault', 'Nissan', 'Ford', 
    'Volkswagen', 'Skoda', 'BMW', 'Mercedes', 'Audi', 'Jaguar', 
    'Land Rover', 'Volvo', 'Porsche', 'Fiat', 'Jeep', 'Mitsubishi'
]

def extract_company(title, page_text):
    """Extract company from title"""
    for brand in CAR_BRANDS:
        if brand.lower() in title.lower():
            return brand
    for brand in CAR_BRANDS:
        if brand.lower() in page_text.lower():
            return brand
    return "Unknown"

def scrape_car_details(link):
    """Scrape one car details using Selenium"""
    try:
        driver.get(link)
        time.sleep(2 + random.random())
        
        page_text = driver.find_element(By.TAG_NAME, "body").text
        
        # Title - gets the full title with year
        try:
            title_elem = driver.find_element(By.TAG_NAME, "h1")
            title = title_elem.text.strip()
        except:
            title = "N/A"
        
        # Company
        company = extract_company(title, page_text)
        
        # Price - look for the large price display
        price = None
        price_patterns = [
            r'₹\s*([\d,]+)',
            r'Rs\.?\s*([\d,]+)'
        ]
        for pattern in price_patterns:
            price_match = re.search(pattern, page_text)
            if price_match:
                price = float(price_match.group(1).replace(',', ''))
                break
        
        # YEAR - Extract from multiple places
        year = None
        
        # Method 1: Look for year in the title first (e.g., "Maruti Suzuki Baleno (2018)")
        year_match_title = re.search(r'\((\d{4})\)', title)
        if year_match_title:
            year = int(year_match_title.group(1))
        
        # Method 2: Look for year in the page text (e.g., "2015-2019" or just "2015")
        if not year:
            year_patterns = [
                r'(\d{4})\s*[–-]\s*\d{4}',  # "2015-2019"
                r'\b(19|20)\d{2}\b'         # "2015" or "2020"
            ]
            for pattern in year_patterns:
                year_match = re.search(pattern, page_text)
                if year_match:
                    # If we matched "2015-2019", take the first year
                    if '-' in year_match.group(0) or '–' in year_match.group(0):
                        year_text = re.search(r'(\d{4})', year_match.group(0))
                        if year_text:
                            year = int(year_text.group(1))
                    else:
                        year = int(year_match.group(0))
                    if 1990 <= year <= 2025:
                        break
                    else:
                        year = None
        
        # KM Driven - Look for number followed by km
        km = None
        km_patterns = [
            r'(\d+[,.]?\d*)\s*(?:km|KM|Km|kms)',
            r'(\d+[,.]?\d*)\s*(?:Kilometers|kilometers)'
        ]
        for pattern in km_patterns:
            km_match = re.search(pattern, page_text, re.IGNORECASE)
            if km_match:
                km_text = km_match.group(1).replace(',', '').replace('.', '')
                if km_text.isdigit():
                    km = float(km_text)
                    break
        
        # Fuel Type - Look for fuel type in the page
        fuel_type = None
        fuel_patterns = {
            'Petrol': r'\b(petrol|gasoline)\b',
            'Diesel': r'\b(diesel)\b',
            'CNG': r'\b(cng)\b',
            'Electric': r'\b(electric|ev)\b',
            'Hybrid': r'\b(hybrid)\b'
        }
        for fuel, pattern in fuel_patterns.items():
            if re.search(pattern, page_text, re.IGNORECASE):
                fuel_type = fuel
                break
        
        # Transmission - Look for transmission type
        transmission = None
        if re.search(r'\b(automatic|auto|amt|cvt)\b', page_text, re.IGNORECASE):
            transmission = 'Automatic'
        elif re.search(r'\b(manual|stick shift)\b', page_text, re.IGNORECASE):
            transmission = 'Manual'
        
        # CITY - Extract ONLY the city, not "Posting date"
        city = "N/A"
        
        # Method 1: Look for "Location" label
        location_match = re.search(r'Location\s*[:\-]?\s*([A-Za-z\s,]+)', page_text, re.IGNORECASE)
        if location_match:
            city = location_match.group(1).strip()
            # Remove extra text if any
            city = re.sub(r'\s*Posting date.*$', '', city)
            city = re.sub(r'\s*Posted.*$', '', city)
        
        # Method 2: Look for city in the breadcrumb or URL
        if city == "N/A" or len(city) < 2:
            # Try to find city from the page structure
            city_patterns = [
                r'in\s+([A-Za-z\s,]+?)(?:\s+[A-Z]|$)',
                r'Location\s*[:\-]?\s*([A-Za-z\s,]+)'
            ]
            for pattern in city_patterns:
                city_match = re.search(pattern, page_text, re.IGNORECASE)
                if city_match:
                    city = city_match.group(1).strip()
                    # Clean up
                    city = re.sub(r'\s*Posting date.*$', '', city)
                    city = re.sub(r'\s*Posted.*$', '', city)
                    if len(city) > 2:
                        break
        
        return {
            'title': title,
            'company': company,
            'price': price,
            'year': year,
            'km_driven': km,
            'fuel_type': fuel_type,
            'transmission': transmission,
            'city': city,
            'link': link
        }
        
    except Exception as e:
        print(f"  Error: {e}")
        return None

def get_links_from_page(page_num):
    """Get all car listing URLs from a specific page"""
    try:
        url = f'https://www.olx.in/cars_c84?page={page_num}'
        print(f"  Page {page_num}: {url}")
        
        driver.get(url)
        time.sleep(3 + random.random())
        
        links = []
        elements = driver.find_elements(By.CSS_SELECTOR, "a[href*='/item/']")
        for elem in elements:
            href = elem.get_attribute('href')
            if href and '/item/' in href:
                links.append(href)
        
        unique_links = list(set(links))
        print(f"  Found {len(unique_links)} listings")
        return unique_links
        
    except Exception as e:
        print(f"  Error: {e}")
        return []

TARGET_CARS = 2100
MAX_PAGES = 1500
MIN_PER_PAGE = 4
MAX_PER_PAGE = 11

all_cars = []
page = 80
failed = 0

print(f"\nTarget: {TARGET_CARS} cars")
print(f"Will take {MIN_PER_PAGE}-{MAX_PER_PAGE} random cars from each page")
print("="*60)

while len(all_cars) < TARGET_CARS and page <= MAX_PAGES:
    print(f"\nPage {page} - Collected {len(all_cars)} cars so far")
    
    page_links = get_links_from_page(page)
    
    if not page_links:
        print("  No links found. Moving to next page...")
        page += 1
        continue
    
    num_to_scrape = random.randint(MIN_PER_PAGE, min(MAX_PER_PAGE, len(page_links)))
    print(f"  Randomly selecting {num_to_scrape} cars from {len(page_links)} available")
    
    selected_links = random.sample(page_links, num_to_scrape)
    
    for i, link in enumerate(selected_links):
        if len(all_cars) >= TARGET_CARS:
            break
            
        print(f"    {i+1}/{num_to_scrape}...", end=" ")
        car_data = scrape_car_details(link)
        
        if car_data and car_data.get('price'):
            all_cars.append(car_data)
            print(f"OK - {car_data['company']} - Rs.{car_data['price']:,.0f} - Year: {car_data['year']} - KM: {car_data['km_driven']} - City: {car_data['city']}")
        else:
            failed += 1
            print(f"FAIL")
    
    if all_cars:
        df_progress = pd.DataFrame(all_cars)
        df_progress.to_csv('car_data_progress.csv', index=False)
        print(f"  Progress saved: {len(all_cars)} cars")
    
    page += 1
    wait = random.randint(3, 8)
    print(f"  Waiting {wait} seconds before next page...")
    time.sleep(wait)

if all_cars:
    df = pd.DataFrame(all_cars)
    
    print(f"\nSummary:")
    print(f"  Total cars scraped: {len(df)}")
    print(f"  Total failed: {failed}")
    
    print(f"\nSample Data (First 10 rows):")
    print(df[['company', 'price', 'year', 'km_driven', 'fuel_type', 'transmission', 'city']].head(10))
    
    print(f"\nMissing Values:")
    print(df.isnull().sum())
    
    # Save final data
    df.to_csv('olx_cars_final_fixed.csv', index=False)
    print(f"\nFinal data saved to 'olx_cars_final_fixed.csv'")
    
    # Show clean sample
    print(f"\nClean Sample:")
    for idx, row in df.head(5).iterrows():
        print(f"  {row['company']} - {row['year']} - {row['km_driven']}km - {row['city']}")
    
else:
    print("No data collected")

driver.quit()
print("\nDone!")
```
